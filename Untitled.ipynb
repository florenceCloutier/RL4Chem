```python
import os
import pickle
import pandas as pd
from rdkit import Chem
from rdkit.Chem import Draw
from PIL import Image
import img2pdf
from reportlab.graphics import renderPDF
from svglib.svglib import svg2rlg
from IPython.display import SVG
from reportlab.lib.pagesizes import letter
from reportlab.pdfgen import canvas
from IPython.display import display, Image
from rdkit import Chem
from rdkit.Chem import Draw
```

```python
for filename in os.listdir('saved/mols'):
    mols = []
    for i,sm in enumerate(pd.read_pickle(f'saved/mols/{filename}').SMILES):
        mols.append(Chem.MolFromSmiles(sm))
        if i == 4:
            break
            
    img = Draw.MolsToGridImage(mols, molsPerRow=5, returnPNG=False)
    img = img.convert('RGB')
    img.save(f'saved/img/{filename[:-13]}.pdf')
    
    print(filename[:-13])
```
